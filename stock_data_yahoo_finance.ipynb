```python
import yfinance as yf
import pandas as pd
```

```python
luv=yf.Ticker("APP")
```

```python
#stock info
luv_info=luv.info
luv_info
```

```python
luv_info["country"]
```

```python
#extracting share prices
luv_price_data=luv.history(period="max")
```

```python
luv_price_data
```

```python
# We can reset the index of the DataFrame with the reset_index function. We also set the inplace paramter to True so the change takes place to the DataFrame itself.
luv_price_data.reset_index(inplace=True)
```

```python
#we can plot the open price against the date
luv_price_data.plot(x="Date",y="Open")
```
